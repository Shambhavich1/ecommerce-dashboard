```python
#monthly revenue trend
df['Date'] = pd.to_datetime(df['Date'])

daily_sales = df.groupby('Date')['Amount'].sum()

plt.figure()
daily_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
#Revenue generated from Top 10 States

state_sales = df.groupby('ship-state')['Amount'].sum().sort_values(ascending=False).head(10)

plt.figure()
state_sales.plot(kind='bar')

plt.title("Top 10 States by Revenue")
plt.xlabel("State")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
#Revenue Contribution from different Categories
sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

cumulative = sales.cumsum() / sales.sum()

plt.bar(sales.index, sales.values)
plt.plot(sales.index, cumulative.values)

plt.xticks(rotation=45)
plt.title("Pareto Analysis - Category Revenue")
plt.show()
```

```python
#Which category performs best in which state

top_states = df.groupby('ship-state')['Amount'].sum().nlargest(10).index

pivot = df[df['ship-state'].isin(top_states)].pivot_table(
    values='Amount',
    index='Category',
    columns='ship-state',
    aggfunc='sum',
    fill_value=0
)

plt.figure()
plt.imshow(pivot, aspect='auto')

plt.colorbar()
plt.xticks(range(len(pivot.columns)), pivot.columns, rotation=90)
plt.yticks(range(len(pivot.index)), pivot.index)

plt.title("Category vs State Sales Heatmap")
plt.show()
```

```python
#Which Category has the highest Cancellations 

cancelled = df[df["Status"] == "Cancelled"]

orders_cancelled = cancelled.groupby("Category")["Order ID"].count()

# Sorting BEFORE plotting
orders_cancelled = orders_cancelled.sort_values(ascending=False)


plt.figure()
orders_cancelled.plot(kind='bar')

plt.title("Cancelled Orders by Category")
plt.xlabel("Category")
plt.ylabel("Number of Cancellations")

plt.xticks(rotation=45)
plt.show()
```

```python
mean_val = df['Amount'].mean()
median_val = df['Amount'].median()

plt.figure(figsize=(10,5))

plt.hist(df['Amount'], bins=40)

# Mean & Median lines
plt.axvline(mean_val, linestyle='dashed', label='Mean')
plt.axvline(median_val, linestyle='solid', label='Median')

plt.title("Revenue Distribution with Mean & Median")
plt.xlabel("Order Value")
plt.ylabel("Frequency")

plt.legend()
plt.tight_layout()

plt.show()
```

```python
#creating a unique Customer Identifier
df['Customer'] = (
    df['ship-city'].astype(str) + "_" +
    df['ship-state'].astype(str) + "_" +
    df['ship-postal-code'].astype(str)
)
```

```python
#Creating Customer-Level Metrics
customer_data = df.groupby('Customer').agg({'Amount': 'sum','Order ID': 'count'                   #total amount spent and number of orders
}).rename(columns={
    'Amount': 'Total_Spend',
    'Order ID': 'Frequency'
})
```

```python
#Defining High vs Low Value Thresholds
freq_median = customer_data['Frequency'].median()
spend_median = customer_data['Total_Spend'].median()
```

```python
high = customer_data[
    (customer_data['Frequency'] >= freq_median) &
    (customer_data['Total_Spend'] >= spend_median)
]

low = customer_data.drop(high.index)
```

```python
plt.figure(figsize=(9,6))

plt.scatter(low['Frequency'], low['Total_Spend'], label='Low Value', alpha=0.6)
plt.scatter(high['Frequency'], high['Total_Spend'], label='High Value', alpha=0.8)

# median lines (quadrants)
plt.axvline(freq_median, linestyle='dashed')
plt.axhline(spend_median, linestyle='dashed')

plt.xlabel("Number of Orders (Frequency)")
plt.ylabel("Total Spending (Amount)")
plt.title("Customer Segmentation (High vs Low Value)")

plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```
